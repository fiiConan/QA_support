```python
import requests
from bs4 import BeautifulSoup

def get_url_input():
    """Prompts the user to enter a URL."""
    while True:
        url = input("請輸入您要檢查的網址 (例如: https://www.google.com): ").strip()
        if url:
            return url
        else:
            print("網址不能為空，請重新輸入。")

def fetch_html(url):
    """Fetches the HTML content from the given URL."""
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        return response.text
    except requests.exceptions.RequestException as e:
        print(f"An error occurred while fetching the URL: {e}")
        return None

def analyze_seo_tags(html_content):
    """Analyzes the HTML content and extracts basic SEO tags."""
    soup = BeautifulSoup(html_content, 'html.parser')
    results = {}

    # Title Tag
    title = soup.find('title')
    results['Title'] = title.get_text(strip=True) if title else 'N/A'

    # Meta Description
    meta_description = soup.find('meta', attrs={'name': 'description'})
    results['Meta Description'] = meta_description.get('content', 'N/A') if meta_description else 'N/A'

    # Meta Keywords
    meta_keywords = soup.find('meta', attrs={'name': 'keywords'})
    results['Meta Keywords'] = meta_keywords.get('content', 'N/A') if meta_keywords else 'N/A'

    # Canonical Link
    canonical_link = soup.find('link', attrs={'rel': 'canonical'})
    results['Canonical URL'] = canonical_link.get('href', 'N/A') if canonical_link else 'N/A'

    # H1 Tags
    h1_tags = [h1.get_text(strip=True) for h1 in soup.find_all('h1')]
    results['H1 Tags'] = h1_tags if h1_tags else 'N/A'

    # Meta Robots
    meta_robots = soup.find('meta', attrs={'name': 'robots'})
    results['Meta Robots'] = meta_robots.get('content', 'N/A') if meta_robots else 'N/A'

    return results

def display_results(results):
    """Prints the extracted SEO tags in a readable format."""
    print("\n--- SEO Tag Analysis Results ---")
    for key, value in results.items():
        print(f"{key}: {value}")
    print("--------------------------------")

# Main execution
if __name__ == "__main__":
    target_url = get_url_input()
    if target_url:
        print(f"Fetching content from: {target_url}")
        html_doc = fetch_html(target_url)
        if html_doc:
            seo_data = analyze_seo_tags(html_doc)
            display_results(seo_data)
        else:
            print("無法獲取網頁內容，請檢查網址或稍後再試。")
    else:
        print("未提供有效的網址。")
```
